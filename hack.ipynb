```python
import moviepy.editor as mp

import speech_recognition as sr
from os import path
from pydub import AudioSegment
import pandas as pd
from googletrans import Translator

from textblob import TextBlob
from gtts import gTTS
```

```python
my_clip = mp.VideoFileClip(r"/content/female voice.mp4")
```

```python
my_clip.audio.write_audiofile(r"my_result.mp3")
```

```python
sound = AudioSegment.from_mp3("my_result.mp3")
sound.export("transcript.wav", format="wav")
AUDIO_FILE = "transcript.wav"

# use the audio file as the audio source
r = sr.Recognizer()
with sr.AudioFile(AUDIO_FILE) as source:
        audio = r.record(source)  # read the entire audio file

        eng_txt = r.recognize_google(audio)

eng_txt
```

```python

blob = TextBlob(eng_txt)
translated_text = blob.translate(from_lang='en', to='hi')

print(translated_text)
```

```python
tts = gTTS(text= str(translated_text) , lang='hi', slow=False)
tts.save('he.mp3')
```

```python
from pydub import AudioSegment

audio1 = AudioSegment.from_file("/content/output/my_result/accompaniment.wav", format="mp3")
audio2 = AudioSegment.from_file("he.mp3", format="mp3")

merged_audio = audio1 * audio2
merged_audio.export("merged_audio.mp3", format="mp3")

print("Audio files merged successfully!")
```

```python
from moviepy.editor import VideoFileClip, AudioFileClip

video = VideoFileClip("/content/female voice.mp4")
audio = AudioFileClip("merged_audio.mp3")

final_video = video.set_audio(audio)
final_video.write_videofile("output_video1.mp4")
```
